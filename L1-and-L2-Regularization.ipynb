
                    #L1 and L2 Regularization

```python
import pandas as pd
import matplotlib.pyplot as  plt
import seaborn as sn
import warnings
warnings.filterwarnings('ignore')
```

We are going to use Melbourne House Price Dataset where we'll predict House Predictions based on various features.


```python
df = pd.read_csv('Melbourne_housing_FULL (1).csv')
df.head()
```

```python
# df.nunique()
```

```python
df.shape
```

```python
uses = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
df = df[uses]
df.head()
```

#Checking for Nan values

```python
df.isna().sum()
```

**Handling Missing values

```python
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
df[cols_to_fill_zero] = df[cols_to_fill_zero].fillna(0)
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
df = pd.get_dummies(df ,drop_first=True)
df.head()
```

```python
# df.CouncilArea_MooraboolShire.Council = df.CouncilArea_MooraboolShire.Council.map({'True': 1, 'False': 0})
```

x= df.drop('Price',axis =1)
y = df['Price']

Let's train our Linear Regression Model on training dataset and check the accuracy on test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y ,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
reg.score(x_train,y_train)
```

#Using Lasso (L1 Regularized)

```python
from sklearn import linear_model
lasso_rag = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_rag.fit(x_train,y_train)
```

```python
lasso_rag.scoreridge_reg.score
```

```python
lasso_rag.score(x_train,y_train)
```

```python
lasso_rag.score(x_test,y_test)
```

#Using Ridge (L2 Regularized) 

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_train,y_train)
```
